## **LECTURA DE DATOS**

```python
import pandas as pd
link='https://github.com/Tracy-12/PRACTICA-2/raw/refs/heads/main/HDI.xlsx'
HDI=pd.read_excel(link)
HDI_2=HDI.copy()
HDI_2.head()
```

## PASO 1

Se elimina las columnas no deseadas

```python
HDI_2=HDI_2.drop(columns=['Back','Unnamed: 3','Unnamed: 5','Unnamed: 7','Unnamed: 9','Unnamed: 11','Unnamed: 12','Unnamed: 13','Unnamed: 14'])
HDI_2.head()
```

Y se cambia el nombre de las columnas

```python
print(HDI_2.columns)
```

```python
HDI_2= HDI_2.rename(columns={'Table 1. Human Development Index and its components ':'Country',
                             'Unnamed: 2': 'Human_Development_Index',
                             'Unnamed: 4': 'Life_expentancy_at_birth',
                             'Unnamed: 6': 'Expected_years_of_schooling',
                             'Unnamed: 8': 'Mean_years_of_schooling',
                             'Unnamed: 10': 'Gross_national_income_per_capita'})
HDI_2
```

### PASO 2

Se identifica las filas sobrantes

```python
HDI_2.iloc[201,0]
```

```python
#Se eilimina a partir de la fila de OTHER COUNTRIEs
HDI_2 = HDI_2.drop(index=range(201, 275))
HDI_2
```

```python
#se busca las filas con datos perdidos
HDI_2[HDI_2.isnull().any(axis=1)]
```

Se elimina las filas no deseadas

```python
HDI_2=HDI_2.dropna(axis='rows',how='any',ignore_index=True)
```

```python
HDI_2.head(192)
```

## PASO 3

```python
HDI_2.info()
```

Se verifica si hay filas con letras

```python
columna=HDI_2['Human_Development_Index'].astype(str)
```

```python
celdas_solo_letras=columna[columna.str.contains(pat=r'[a-zA-Z]',regex=True)]
celdas_solo_letras
```

```python
#se elimina la fila
HDI_2=HDI_2.drop(index=0)
HDI_2
```

## PASO 4

```python
HDI_2.info()
```

Como los datos numericos siguen como object , se convierte los datos

```python
#PRIMERA COLUMNA
HDI_2['Human_Development_Index']=pd.to_numeric(HDI_2['Human_Development_Index'])
HDI_2.info()
```

```python
#SEGUNDA COLUMNA
HDI_2['Life_expentancy_at_birth']=pd.to_numeric(HDI_2['Life_expentancy_at_birth'])
HDI_2.info()
```

```python
#TERCERA COLUMNA
HDI_2['Expected_years_of_schooling']=pd.to_numeric(HDI_2['Expected_years_of_schooling'])
HDI_2.info()
```

```python
#CUARTA COLUMNA
HDI_2['Mean_years_of_schooling']=pd.to_numeric(HDI_2['Mean_years_of_schooling'])
HDI_2.info()
```

```python
#QUINTA COLUMNA
HDI_2['Gross_national_income_per_capita']=pd.to_numeric(HDI_2['Gross_national_income_per_capita'])
HDI_2.info()
```

## PASO 5

Por ultimo se aade la columna con la categorias

```python
HDI_2['Category']=pd.cut(HDI_2['Human_Development_Index'],
                         bins=[0,0.549,0.699,0.796,1],
                         labels=['LOW HUMAN DEVELOPMENT','MEDIUM HUMAN DEVELOPMENT','HIGH HUMAN DEVELOPMENT','VERY HIGH HUMAN DEVELOPMENT'],
                           ordered=True)
HDI_2
```

```python
HDI_2['Human_Development_Index'] = HDI_2['Human_Development_Index'].apply(lambda x: round(x, 3))
HDI_2['Life_expentancy_at_birth'] = HDI_2['Life_expentancy_at_birth'].apply(lambda x: round(x, 1))
HDI_2['Expected_years_of_schooling'] = HDI_2['Expected_years_of_schooling'].apply(lambda x: round(x, 1))
HDI_2['Mean_years_of_schooling'] = HDI_2['Mean_years_of_schooling'].apply(lambda x: round(x, 1))
HDI_2['Gross_national_income_per_capita'] = HDI_2['Gross_national_income_per_capita'].apply(lambda x: round(x, 0))
HDI_2
```

```python
HDI_2.to_excel('HDI_2.xlsx')
```
